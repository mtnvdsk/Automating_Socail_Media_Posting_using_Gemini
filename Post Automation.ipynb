```
from IPython.display import Markdown

model = genai.GenerativeModel('gemini-pro')
```

```
topic=input()
#response = model.generate_content(Prompt)
topic_prompt=f"""Write a article on the {topic} of 3000 characters using the following steps 
steps to write the article
Do not use any special characters in the article
1.Introduction to the topic in a nice and depth of the topic in 5 lines or 100 characters.
2.Explain the topic with a simple example of the real world scenario.
3.Write a code if any for the topic.
4.Explain the time and space complexity of the code if any written in the previous step
5.Write a conclusion of the topic.

Generate in a plain text to be able to post on social media"""
generate_article= model.generate_content(topic_prompt,generation_config=genai.types.GenerationConfig(
        candidate_count=1,
        stop_sequences=[],
        max_output_tokens=400,
        temperature=0.7))
article=generate_article.text
titleprompt=f"Write a title for this article {generate_article.text} in 10 words"
Title= model.generate_content(titleprompt)
title_generated=Title.text
```

```
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont

# Needed for compatibility with some systems where cv2 can't find libGL
import os
os.environ['OPENCV_IO_ENABLE_JASPER']= 'true'
os.environ['OPENCV_IO_ENABLE_IMAGEIO']= 'true'
os.environ['OPENCV_IO_ENABLE_GDAL']= 'true'
os.environ['OPENCV_IO_ENABLE_OPENEXR']= 'true'

def create_colored_rectangle(image_path, x, y, text):
    # Load the image
    image = cv2.imread(image_path)
    
    # Get the dimensions of the image
    height, width, _ = image.shape
    
    # Calculate the coordinates for the rectangle
    top_left = (int((width - x) / 2), int((height - y) / 2))
    bottom_right = (top_left[0] + x, top_left[1] + y)
    
    # Create a colored rectangle
    color = (0, 0, 0)  # RGB color (here, green)
    cv2.rectangle(image, top_left, bottom_right, color, thickness=cv2.FILLED)
    
    # Convert the image to PIL format to add text
    pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    draw = ImageDraw.Draw(pil_image)
    
    # Set the font for the text
    font_path = "sample.ttf"  # Replace with the path to a TTF font file
    font_size = 10
    font = ImageFont.truetype(font_path, font_size)
    
    # Calculate the position to center the text on the rectangle
    text_width, text_height = draw.textsize(text, font)
    text_position = ((width - text_width) // 2, (height - text_height) // 2)
    
    # Add text to the image
    draw.text(text_position, text, font=font, fill=(255, 255, 255))  # RGB color (here, white)
    
    # Save or display the result
    pil_image.save("output_image.jpg")  # Save the image
    pil_image.show()  # Display the image

# Replace 'your_image.jpg' with the path to your input image
create_colored_rectangle('sample.png', 100, 20, 'VOICE')
```
